```python
import pandas as pd
import geopandas as gpd
```

```python
# Add shapefile of new affordable housing projects in LA 
lahousing = gpd.read_file('LAHD_Affordable_Housing_Projects_List__2003_to_Present_.csv')
```

```python
# How many affordable housing projects for neighborhood
lahousing['SITE COMMUNITY'].value_counts()
```

```python
# Saved the column Site Community as a variable
housing_count = lahousing['SITE COMMUNITY'].value_counts()
housing_count
```

```python
# Converted to dataframe
housing_count = housing_count.reset_index()
housing_count
```

```python
# Renamed columns
housing_count.columns = ['Site Community', 'Affordable Housing Project Count']
```

```python
housing_count
```

```python
# Plotted the housing count data by plot and bar graph
housing_count.plot()
```

```python
housing_count.plot.bar()
```

```python
#Trimmed data to only show neighborhood, housing type, and the projects total units
desired_columns = ['SITE COMMUNITY','HOUSING TYPE','PROJECT TOTAL UNITS']
```

```python
# Showing a sample of the trimmed data
lahousing[desired_columns].sample(5)
```

```python
lahousing_trimmed = lahousing[desired_columns].copy()
```

```python
# Displaying trimmed data
lahousing_trimmed
```
